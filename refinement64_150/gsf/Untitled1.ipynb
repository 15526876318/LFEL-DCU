```python
import matplotlib.pyplot as plt 
import numpy as np 

def get_train_val_info(f1):
    train = []
    val = []
    acc_t = []
    acc_v = []
    with open (f1,'r') as t:
        lines = t.readlines()
        for line in lines:
            try:
                a,_,loss,acc = line.split(" ")
            except:
                continue
            #print(a)
            if a =="train:":
                train.append(np.round(float(loss),4))
                acc_t.append(np.round(float(acc),4))
            elif a == "val:":
                val.append(np.round(float(loss),4))
                acc_v.append(np.round(float(acc),4))
    return train,val,acc_t,acc_v

def plot_loss_acc(train,val,acc_t,acc_v):
    font = 20
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(13,5))
    ax = axes.flatten()
    ax[0].plot( list(range(len(train))),train,label = 'train', color='blue',linewidth=1.0,linestyle='-')
    ax[0].plot( list(range(len(train))),val,label = 'validation', color='red',linewidth=1.0,linestyle='-')
    ax[1].plot( list(range(len(train))),acc_t,label = 'train', color='blue',linewidth=1.0,linestyle='-')
    ax[1].plot( list(range(len(train))),acc_v,label = 'validation', color='red',linewidth=1.0,linestyle='-')
    ax[0].set_ylabel("Loss", fontsize=font)
    ax[1].set_ylabel("ACC", fontsize=font)
    ax[0].set_xlabel("EPOCH", fontsize=font)
    ax[1].set_xlabel("EPOCH", fontsize=font)
    ax[0].legend(fontsize=font)
    ax[1].legend(fontsize=font)
    plt.savefig("loss.jpg")
    plt.show()
```

```python
f1 = "train_seq_dist_angle_sincos16active_mish_770.log"
train,val,acc_t,acc_v = get_train_val_info(f1)
plot_loss_acc(train,val,acc_t,acc_v)
```
